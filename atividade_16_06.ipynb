```python
#Solicite ao usuário um número inteiro positivo.

#Utilizando um laço de repetição, calcule e exiba a soma de
#todos os números de 1 até o número informado.

#Exemplo:

#Digite um número: 5

#Resultado: 15

#Pois:

#1 + 2 + 3 + 4 + 5 = 15
a = int(input("digite um numero "))
for i in range(1,a):
    a += i
print(a)
```

```python
# Solicite ao usuário um número inteiro e exiba sua tabuada
# de 1 até 10.

# Exemplo:

# Digite um número: 7

# 7 x 1 = 7
# 7 x 2 = 14
# 7 x 3 = 21
# ...
# 7 x 10 = 70

a = int(input("digite um numero "))
print(a * 1)
print(a * 2)
print(a * 3)
print(a * 4)
print(a * 5)
print(a * 6)
print(a * 7)
print(a * 8)
print(a * 9)
print(a * 10)
```

```python
#Solicite ao usuário um número inteiro positivo e calcule
#seu fatorial utilizando um laço de repetição.

#Exemplo:

#Digite um número: 5

#Fatorial = 120

#Pois:

#5! = 5 × 4 × 3 × 2 × 1

#Observação:
#O programa deve validar se o número informado é positivo.

a = int(input("digite um numero "))
try:
  for i in range(1,a):
    a = a * i
  print(a )
except a >= 0:
  print("digite um numero positivo")
```
