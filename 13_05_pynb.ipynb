```python
# 1- Contagem simples, o usuário digita um número e o programa imprime do zero até o número

num = int(input("Digite um número: "))

for i in range(num):
  print(i + 1)
```

```python
# 2- Contagem regressiva, o usuário digita um número e o progresso e o programa imprime do número até zero

num = int(input("Digite um número: "))

for i in range(num + 1):
  print(num - i)
```

```python


# 3- Números pares, o usuário digita um número e o programa imprime os números pares

num = int(input("Digite um número: "))

for i in range(num + 1):
  if i % 2 == 0:
    print(i)
```

```python
# 4- Números impares, o usuário digita um número e o programa imprime os números impares

num = int(input("Digite um número: "))

for i in range(num + 1):
  if i % 2 == 1:
    print(i)
```

```python
# 5- Soma, o usuário digita um número e o programa soma os números do zero até o número

num = int(input("Digite um número: "))

soma = 0
for i in range(num + 1):
  soma += i

print(f"A soma dos números de 0 até {num} é: {soma}")
```

```python
# 6- Tabuada, o usuário digita um número e o programa imprime a tabuada do número

num = int(input("Digite um número: "))

for i in range(11):
  print(f"{num} X {i} = {num * i}")
```

```python
# 7- Contar letras, o usuário digita uma palavra e o programa conta as letras da palavra

palavra = str(input("Digite uma palavra: "))
letras = 0
for i in (palavra):
  letras += 1
print (letras)
```

```python
# 8- Contar vogais, o usuário digita uma palavra e o programa conta as vogais

palavra = input("Digite uma palavra: ").lower()
vogais = "aeiou"
contador_vogais = 0

for letra in palavra:
    if letra in vogais:
        contador_vogais += 1

print(f"A palavra '{palavra}' possui {contador_vogais} vogais.")
```

```python
# 9- Maior número digitado, o usuário digita 4 números e o programa imprime o maior número digitado

num1 = int(input("Digite um número: "))
num2 = int(input("Digite um número: "))
num3 = int(input("Digite um número: "))
num4 = int(input("Digite um número: "))

if num1 > num2 and num1 > num3 and num1 > num4:
  print(f"O maior número é {num1}")

if num2 > num1 and num2 > num3 and num2 > num4:
  print(f"O maior número é {num2}")

if num3 > num1 and num3 > num2 and num3 > num4:
  print(f"O maior número é {num3}")

if num4 > num2 and num4 > num3 and num4 > num1:
  print(f"O maior número é {num4}")
```

```python
# 10- O usuário digita um número entre 01 e 100 e o programa mostra quais são divisiveis por 3 e 5
print("Números entre 1 e 100 divisíveis por 3 e 5:")
for num in range(1, 101):
    if num % 3 == 0 and num % 5 == 0:
        print(num)
```

```python
#11 piramide numerica simetrica
# faça um programa que receba um numero N e imprima uma piramide perfeitamente centraliza
#exemplo para N = 5 :
#     1
#    212
#   32123
#  4321234
# 543212345

n = int(input("Digite um número para o topo da pirâmide (N): "))

for i in range(1, n + 1):
    # Adiciona os espaços em branco para centralizar
    print(" " * (n - i), end="")

    # Parte decrescente da pirâmide (ex: 321)
    for j in range(i, 0, -1):
        print(j, end="")

    # Parte crescente da pirâmide (ex: 23)
    for k in range(2, i + 1):
        print(k, end="")

    print()
```
