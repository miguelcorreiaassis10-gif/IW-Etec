```python
saldo = 1000.00

print("Bem-vindo ao Caixa Eletrônico!")

while True:
    print("--- Menu --- ")
    print("1. Ver Saldo")
    print("2. Depositar Dinheiro")
    print("3. Sacar Dinheiro")
    print("4. Sair")

    try:
        opcao = int(input("Escolha uma opção: "))

        if opcao == 1:
            print(f"Seu saldo atual é: R$ {saldo:.2f}")
        elif opcao == 2:
            valor_deposito = float(input("Digite o valor a depositar: "))
            if valor_deposito > 0:
                saldo += valor_deposito
                print(f"Depósito de R$ {valor_deposito:.2f} realizado com sucesso.")
                print(f"Novo saldo: R$ {saldo:.2f}")
            else:
                print("O valor do depósito deve ser positivo.")
        elif opcao == 3:
            valor_saque = float(input("Digite o valor a sacar: "))
            if valor_saque > 0:
                if saldo >= valor_saque:
                    saldo -= valor_saque
                    print(f"Saque de R$ {valor_saque:.2f} realizado com sucesso.")
                    print(f"Novo saldo: R$ {saldo:.2f}")
                else:
                    print("Saldo insuficiente.")
            else:
                print("O valor do saque deve ser positivo.")
        elif opcao == 4:
            print("Obrigado por usar nosso Caixa Eletrônico. Volte sempre!")
            break
        else:
            print("Opção inválida. Por favor, escolha uma opção de 1 a 4.")
    except ValueError:
        print("Entrada inválida. Por favor, digite um número para a opção.")

Bem-vindo ao Caixa Eletrônico!
--- Menu ---
1. Ver Saldo
2. Depositar Dinheiro
3. Sacar Dinheiro
4. Sair
Escolha uma opção: 1
Seu saldo atual é: R$ 1000.00
--- Menu ---
1. Ver Saldo
2. Depositar Dinheiro
3. Sacar Dinheiro
4. Sair
Escolha uma opção: 2
Digite o valor a depositar: 10
Depósito de R$ 10.00 realizado com sucesso.
Novo saldo: R$ 1010.00
--- Menu ---
1. Ver Saldo
2. Depositar Dinheiro
3. Sacar Dinheiro
4. Sair
Escolha uma opção: 3
Digite o valor a sacar: 1100
Saldo insuficiente.
--- Menu ---
1. Ver Saldo
2. Depositar Dinheiro
3. Sacar Dinheiro
4. Sair
Escolha uma opção: 3
Digite o valor a sacar: 1000
Saque de R$ 1000.00 realizado com sucesso.
Novo saldo: R$ 10.00
--- Menu ---
1. Ver Saldo
2. Depositar Dinheiro
3. Sacar Dinheiro
4. Sair
Escolha uma opção: 2
Digite o valor a depositar: 990
Depósito de R$ 990.00 realizado com sucesso.
Novo saldo: R$ 1000.00
--- Menu ---
1. Ver Saldo
2. Depositar Dinheiro
3. Sacar Dinheiro
4. Sair
Escolha uma opção: 4
Obrigado por usar nosso Caixa Eletrônico. Volte sempre!
```

```python
# Urna Eletrônica

# Criar um programa de votação com 3 candidatos

# Votar candidato A
# Votar candidato B
# Votar candidato C
# Votar Branco
# Votar Nulo
# Encerrar votação

# O sistema deve repetir até encerrar a votação
# Contar votos de cada candidato
# Contar votos brancos e nulos
# Ao encerrar, mostrar:

# Total de Votos
# Total de votos por candidato
# Total votos brancos
# Total votos nulos
# Vencedor da eleição

print("bem vindo")
ca = 0
cb = 0
cc = 0
branco = 0
nulo = 0
ab = 0
try:
  while  ab != 10:
    voto = str(input(" digite o nome do seu candidato "))
    if voto == "a":
      print("voto computado")
      ca = ca + 1
      ab = ab + 1
    elif voto == "b":
          print("voto computado")
          cb = cb + 1
    elif voto == "c":
          print("voto computado")
          cc = cc + 1
    elif voto == "branco":
          print("voto computado")
          branco = branco + 1
    elif voto == "nulo":
          print("voto computado")
          nulo = nulo + 1
    else :
          print("encerrando")
          break
    if ab == 10:
      print("total de votos", ab)
      print("total de votos do candidato a", ca)
      print("total de votos do candidato b", cb)
      print("total de votos do candidato c", cc)
      print("total de votos brancos", branco)
      print("total de votos nulos", nulo)
      if ca > cb and ca > cc:
          print("o vencedor é o candidato a")
      elif cb > ca and cb > cc:
          print("o vencedor é o candidato b")
      else:
          print("o vencedor é o candidato c")
except:
      print("coloque um nome elegivel")
```
