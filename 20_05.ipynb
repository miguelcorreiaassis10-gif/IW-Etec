```python
# escreva um programa que receba 4 notas de um aluno, calcula a media e imprime aprovado, reculperação ou reprovado
#regras: utilizar while, nota >= 7 aprovado, maior ou igual a 5, menmor de 5 reculperaçao

nome = str(input("qual seu nome? "))


bim = 1

while bim < 4:
  if bim == 1:
    n1 = float(input("nota primeiro bimmestre "))
  if bim == 2:
    n2 = float(input("nota segundo bimestre  "))
  elif bim == 3:
    n3 = float(input("nota terceiro bimestre "))
  else:
    n4 = float(input("nota quarto bimestre "))
  bim += 1


nf = n1 + n2 + n3 + n4
nb = nf / 4
if n1 > 10 or n2 > 10 or n3 > 10 or n4 > 10:
  print("digite notas validas (0 a 10)")
elif nb >= 7:
  print(f"o aluno {nome}  tirou {nb}  e foi aprovado")
elif nb >= 5:
    print(f"o aluno {nome} tirou {nb} e ficou de recuperação")
else:
    print(f"o aluno {nome} tirou {nb} foi reprovado")
```

```python
# voce foi contratado para desenvolver um sistema simples de caixa eletronico
# o sistema devera funcionar atraves de um menu interativo utilizando a estrutura de repetição de while
# o programa com saldo inicial de R$ 1.000,00 e permitir que o usuario escolha diferentes operações até decidir do sistema
# funcionalidades
# 1 - ver saldo
# 2 - depositar  dinheiro
# 3 - sacar dinheiro
# 4 - sair
# regras:
# ver saldo
# Ao escolher essa opção, o sistema deverá mostrar o saldo atual
#depositar dinheiro
#usuario

```

```python
saldo = 1000.00

print("Bem-vindo ao Caixa Eletrônico!")

while True:
    print("--- Menu --- ")
    print("1. Ver Saldo")
    print("2. Depositar Dinheiro")
    print("3. Sacar Dinheiro")
    print("4. Sair")

    try:
        opcao = int(input("Escolha uma opção: "))

        if opcao == 1:
            print(f"Seu saldo atual é: R$ {saldo:.2f}")
        elif opcao == 2:
            valor_deposito = float(input("Digite o valor a depositar: "))
            if valor_deposito > 0:
                saldo += valor_deposito
                print(f"Depósito de R$ {valor_deposito:.2f} realizado com sucesso.")
                print(f"Novo saldo: R$ {saldo:.2f}")
            else:
                print("O valor do depósito deve ser positivo.")
        elif opcao == 3:
            valor_saque = float(input("Digite o valor a sacar: "))
            if valor_saque > 0:
                if saldo >= valor_saque:
                    saldo -= valor_saque
                    print(f"Saque de R$ {valor_saque:.2f} realizado com sucesso.")
                    print(f"Novo saldo: R$ {saldo:.2f}")
                else:
                    print("Saldo insuficiente.")
            else:
                print("O valor do saque deve ser positivo.")
        elif opcao == 4:
            print("Obrigado por usar nosso Caixa Eletrônico. Volte sempre!")
            break
        else:
            print("Opção inválida. Por favor, escolha uma opção de 1 a 4.")
    except ValueError:
        print("Entrada inválida. Por favor, digite um número para a opção.")
```
