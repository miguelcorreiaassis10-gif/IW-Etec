```python
# decisões

# entrada
a = int(input("digite um numero: "))
b = int(input("digite outro numero: "))
#processamento e saida
if a > b:
  print(f"{a} é maior que {b}")
elif a == b:
  print(f"{a} é igual a {b}")
else:
  print(f"{b} é maior que {a}")
```

escrever um programa que recebe 4 notas do aluno e calcula a media para mostrar que foi aprovado ou reprovado

```python
#entrada
nota1 = float (input("sua nota 1 bimestre "))
nota2 = float (input("sua nota 2 bimestre "))
nota3 = float (input("sua nota 3 bimestre "))
nota4 = float (input("sua nota 4 bimestre "))
#processamento
notaF = nota1 + nota2 + nota3 + nota4
notaB = notaF / 4
print(f"sua nota bimestral é :{notaB}")
#saida
if notaB >= 7.0:
  print("aprovado ")
elif notaB >= 5.0:
  print("recuperação ")
else:
  print("reprovado ")
```

```python
print("--------------------MENU--------------------")
print("")
print("opção - 1 = SOMA")
print("opção - 2 = SUBTRAÇÃO")
print("opção - 3 = MULTIPLICAÇÃO")
print("opção - 4 = DIVISÃO")
print("")
opcao = int(input("digite a OPÇÃO "))
if opcao <= 0 or opcao > 4:
  print("OPÇÃO INVALIDA")
#entrada:
a = float(input("digite um numero: "))
b = float(input("digite outro numero:"))
#processamento e saida
if opcao == 1:
  print("------------SOMA--------------")
  print(f"a soma é {a + b}")
elif opcao == 2:
  print("------------SUBTRAÇÃO-------------")
  print(f"a subtração é {a - b}")
elif opcao == 3:
  print("-------------MULTIPLICAÇÃO-------------")
  print(f"a multiplicação é {a * b}")
elif opcao == 4:
  print("------------DIVISÃO------------")
  print(f"a divisão é {a / b}")
else:
  print("OPÇÃO INVALIDA")
```
