faça um codigo para ver se o numero é par ou impar

```python
numero = int(input("digite um numero: "))
num = numero % 2
if num == 0:
  print(f"o numero {numero} é par")
else:
  print(f" o numero {numero} é impar")
```

faça um codigo que recebe o nome e 4 notas referente aos bimestres,calcula a media e mostra se passou, recuperação ou reprovado

```python
nome = str(input("qual seu nome? "))
n1 = float(input(" qual sua nota do Primeiro bimestre? "))
if n1 > 10:
  print(" digite uma nota valida  (de 0 a 10)")
n2 = float(input(" qual sua nota do Segundo bimestre? "))
if n2 > 10:
  print(" digite uma nota valida  (de 0 a 10)")
n3 = float(input(" qual sua nota do Terceiro bimestre? "))
if n3 > 10:
  print(" digite uma nota valida  (de 0 a 10)")
n4 = float(input(" qual sua nota do Quarto bimestre? "))
if n4 > 10:
  print(" digite uma nota valida  (de 0 a 10)")
nf = n1 + n2 + n3 + n4
nb = nf / 4
if n1 > 10 or n2 > 10 or n3 > 10 or n4 > 10:
  print("digite notas validas (0 a 10)")
elif nb >= 7:
  print(f"o aluno {nome}  tirou {nb}  e foi aprovado")
elif nb >= 5:
    print(f"o aluno {nome} tirou {nb} e ficou de recuperação")
else:
    print(f"o aluno {nome} tirou {nb} foi reprovado")
```

#escreva um programa que recebe os valores de 3 lados para verificar se um triangulo é valido
classificar o tipo de triangulo

```python
#Equilátero: Três lados e três ângulos internos iguais (todos de 60 graus).
#Isósceles: Dois lados iguais e os ângulos da base também iguais.
#Escaleno: Três lados e três ângulos com medidas diferentes.
# a soma de quaisquer dois lados deve ser sempre maior que o terceiro lado
l1 = float(input("digite o primeiro lado"))
l2 = float(input("digite o segundo lado"))
l3 = float(input("digite o terceiro lado"))
if  l1 + l2 > l3 and l3 + l2 > l1 and l3 + l1 > l2:
  print(" é um triangulo valido")
else:
  print(" é um triangulo invalido")

if l1 == l2 and l2 == l3:
  print("é um triangulo Equilátero ")
elif  l2 == l1 or l3 == l2 or l3 == l1:
 print("é um triangulo Isósceles ")
else:
  print("é um triangulo Escaleno ")
```
