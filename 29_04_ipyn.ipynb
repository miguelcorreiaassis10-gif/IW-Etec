```python
# escrever um programa que o usuario digita uma palavra e o programa verifique se é uma palindroma e imprimir
palavra = str(input("digite uma palavra "))
invertida = ""
for i in (palavra):
  invertida = i + invertida
if invertida == palavra:
    print(f"a palavra {palavra} invertida é {invertida} é uma palindroma ")
else:
    print(f"a palavra {palavra} invertida é {invertida} não é uma palindroma ")
```

```python
# soma acumula peça ao usario para digitar um numero e imprimi o resultado da soma do 1 até o numero digitado
n = int(input("digite um numero "))
for i in range(1,n +1,1):
  print(i)
```

```python
# peça um numero ao usuario e o programa irá separar os numeros pares e impares
# pares : 0,2,4,6,8
# inpares:1,3,5,7,9
a = int(input("digite um numero "))
print("------------------------PARES-----------------------")
for i in range(0,a,2):
    print(i)
print("------------------------IMPARES---------------------")
for i in range(1,a,2):
    print(i)
```
