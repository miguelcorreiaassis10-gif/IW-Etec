```python
# Cofre digital

# Um cofre possui um código composto por 5 números inteiros.

# Desenvolva um programa que:

# Leia os 5 números utilizando um laço de repetição.

# Ao final informar:

# 1 - Quantos números são pares.
# 2 - Quantos números são ímpares.
# 3 - O maior número.
# 4 - A soma de todos os números.
# 5 - A média dos números.
# 6 - O menor número.
# 7 - Quantos números são maiores que a média.

numbers = []
pares = 0
impares = 0
soma = 0
maior_numero = -float('inf')
menor_numero = float('inf')

print('Por favor, digite 5 números inteiros:')

for i in range(5):
    while True:
        try:
            num = int(input(f'Digite o {i+1}º número: '))
            numbers.append(num)
            soma += num

            if num % 2 == 0:
                pares += 1
            else:
                impares += 1

            if num > maior_numero:
                maior_numero = num

            if num < menor_numero:
                menor_numero = num
            break
        except ValueError:
            print('Entrada inválida. Por favor, digite um número inteiro.')

media = soma / len(numbers) if numbers else 0

maiores_que_media = 0
for num in numbers:
    if num > media:
        maiores_que_media += 1

print('\n--- Resultados ---')
print(f'1 - Quantos números são pares: {pares}')
print(f'2 - Quantos números são ímpares: {impares}')
print(f'3 - O maior número: {maior_numero}')
print(f'4 - A soma de todos os números: {soma}')
print(f'5 - A média dos números: {media:.2f}')
print(f'6 - O menor número: {menor_numero}')
print(f'7 - Quantos números são maiores que a média: {maiores_que_media}')
```
